```python
#import packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#prepare data
x = [1,2,3,4,5,6]
y = [9,8,7,6,5,4]

x2 = np.arange(0,5,.5)

plt.plot(x2[:9],x2[:9]**2, color= 'blue', linestyle= '-.')

plt.plot(x,y, label = 'negative correlation', color = 'red', linewidth = 2, linestyle= '--', marker= '*',)
plt.xlabel('x - label', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20, 'fontweight': 'bold'})
plt.ylabel('y - label')
plt.title('Created my first python graph!')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

gas = pd.read_csv('../Activities/gas_prices.csv')
#print(gas)

x = (gas['Year'])
y = gas.USA 
w = gas.Canada
v = gas.UK

plt.plot(x,y, label= "USA")
plt.plot(x,w, label = 'Canada')
plt.plot(x,v, label="UK")

plt.xticks(x[::3])
plt.ylabel('Price per Gallon')
plt.xlabel('Year')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

gas = pd.read_csv('../Activities/gas_prices.csv')
print(gas.head())

x = (gas['Year'])
countries = ['Australia', 'USA', 'UK', 'Canada','France', 'Germany', 'Italy', 'Japan', 'Mexico', 'South Korea']
for country in gas:
    if country in countries:
        plt.plot(x, gas[country])


plt.xticks(x[::3])
plt.ylabel('Price per Gallon')
plt.xlabel('Year')
#plt.legend()
plt.show()
```
