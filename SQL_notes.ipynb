--5. The EV charging firm is wondering which charging Stations (stationId) are being used the least. They would like to move these stations to a place with a greater need. Please find the total number of hours (`sumTotalHrs`) that each station is used. The output should contain all of the stations that have a total usage of less than 5 hours. 

SELECT stationId,
sum(chargeTimeHrs) as 'sumTotalHrs'
FROM EVCharge
GROUP BY stationId
HAVING sum(chargeTimeHrs) < 5;

-- 2. Using dimUser and factCharge, which app platform had the most amount of charging sessions. Return `appPlatform` and `numCharges`. Order the result set from highest to lowest number of charges. 

SELECT u.appPlatform as 'appPlatform',
count(*) as 'numCharges'
FROM factCharge f
INNER Join dimUser u
ON f.userId = u.userId
GROUP BY u.appPlatform
ORDER BY 2 DESC;
