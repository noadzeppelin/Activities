1. Use the given x and y axis values to create a line graph
    - other instuctions are written in as comments 

```python
#import Packages 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# x and y axis values
x = [1,2,3,4,5,6]
y = [2,4,1,5,2,6]

# plotting the points
    # change the line color to greeen
    # change line style to dashed 
    # change linewidth to 3
    # change the marker style (use documentation and pick one)
    # change marker color to blue 
    # change marker size to 12
plt.plot(x,y, color= 'green', linestyle= '--', linewidth= 3, marker='*', markerfacecolor= "b",markersize= 12)

#tweak  
    # setting x and y axis range to 0,8 using x/y lim function
plt.xlim(0,8)
plt.ylim(0,8)

    # naming the x  and y axis
plt.xlabel('This is the X-axis')
plt.ylabel('This is the Y-axis')

    # add a  a title to my graph
plt.title('Spikey Graph')

# render plot 
plt.show()
```
